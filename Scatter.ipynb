```python
import pandas as pd
```

```python
df = pd.read_csv("/content/pinguinos.csv")
df
```

```python
# Seleccionar solo las columnas numÃ©ricas
numeric_cols = df.select_dtypes(include=['float', 'int']).columns
# Reemplazar los valores faltantes en las columnas numÃ©ricas por la media
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
df
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
ax = df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm',
    title="TamaÃ±o de aleta vs TamaÃ±o de pico",

)
ax.set_xlabel("TamaÃ±o de aleta")
ax.set_ylabel("TamaÃ±o de pico")
ax.figure.savefig("scatter.png")
```

```python
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm',
    c="body_mass_g"
)
```

```python
import plotly.express as px
px.scatter(df, x="flipper_length_mm", y="culmen_length_mm", color="species")
```

```python
import seaborn as sns
ax = sns.catplot(
    data=df,
    y="species",
    x="flipper_length_mm",
    hue="island"
)
ax.set_axis_labels("TamaÃ±o de aleta", "Especie")
ax.figure.savefig("catplot.png")
```
